## 2.2 実データを使った生成モデルの実装

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_name = 'height'
path = os.path.join(data_name + '.txt')
xs = np.loadtxt(path)
print(xs.shape)

plt.hist(xs, bins=100, density=True)
plt.xlabel('Height[cm]')
plt.ylabel('Probability Density')
plt.title('Histogram of Height')
plt.show()
```

```python
mu = np.mean(xs)
sigma = np.std(xs)

print('mean:', mu)
print('std:', sigma)
```

```python
def normal(x, mu=0, sigma=1):
    return np.exp(-0.5 * ((x - mu) / sigma) ** 2) / (np.sqrt(2 * np.pi) * sigma)

x = np.linspace(150, 190, 1000)
y = normal(x, mu, sigma)

plt.hist(xs, bins=100, density=True)
plt.plot(x, y, color='red')
plt.xlabel('Height[cm]')
plt.ylabel('Probability Density')
plt.title('Histogram of Height')
plt.show()
```

## 2.4 生成モデルの用途

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_name = 'height'
path = os.path.join(data_name + '.txt')
xs = np.loadtxt(path)

mu = np.mean(xs)
sigma = np.std(xs)
samples = np.random.normal(mu, sigma, 10000)

plt.hist(xs, bins=100, density=True, label='orginal')
plt.hist(samples, bins=100, density=True, alpha=0.5, label='generated')
plt.xlabel('Height[cm]')
plt.ylabel('Probability Density')
plt.title('Histogram of Height')
plt.legend()
plt.show()
```

```python
from scipy.stats import norm

x = 1.0
p = norm.cdf(x, loc=0, scale=1)

print(p)
```

```python
data_name = 'height'
path = os.path.join(data_name + '.txt')
xs = np.loadtxt(path)

mu = np.mean(xs)
sigma = np.std(xs)

p1 = norm.cdf(160, loc=mu, scale=sigma)
print(f"p(x <= 160) = {p1}")

p2 = norm.cdf(180, loc=mu, scale=sigma)
print(f"p(x > 180) = {1 - p2}")
```
