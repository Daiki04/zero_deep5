## 1.2 正規分布

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normal(x, mu=0, sigma=1):
    return 1/(np.sqrt(2*np.pi)*sigma) * np.exp(-0.5*((x-mu)**2) / sigma**2)
```

```python
x = np.linspace(-5, 5, 100)
y = normal(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Normal distribution with mu=0, sigma=1")
plt.show()
```

```python
colors = ['r', 'g', 'b']
mus = [3, 0, 5]
x = np.linspace(-5, 10, 100)

for i, mu in enumerate(mus):
    y = normal(x, mu=mu)
    plt.plot(x, y, label=f'mu={mu}', c=colors[i])

plt.xlabel('x')
plt.ylabel('y')
plt.title("Normal distribution with different mu")
plt.legend()
plt.show()
```

```python
colors = ['r', 'g', 'b']
sigmas = [0.5, 1, 2]
x = np.linspace(-10, 10, 100)

for i, sigma in enumerate(sigmas):
    y = normal(x, sigma=sigma)
    plt.plot(x, y, label=f'sigma={sigma}', c=colors[i])

plt.xlabel('x')
plt.ylabel('y')
plt.title("Normal distribution with different sigma")
plt.legend()
plt.show()
```

## 1.3 中心極限定理

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 3 # サンプルサイズ

xs = []

for i in range(N):
    x = np.random.rand() # 一様乱数を生成
    xs.append(x)

x_mean = np.mean(xs)
print(f"平均: {x_mean}")
```

```python
x_means = []
N = 1

for _ in range(10000):
    xs = []
    for i in range(N):
        x = np.random.rand() # 一様乱数を生成
        xs.append(x)
    x_mean = np.mean(xs)
    x_means.append(x_mean)

plt.hist(x_means, bins=100, range=(0, 1), density=True)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title(f"N = {N}")
plt.xlim(-0.05, 1.05)
plt.ylim(0, 5)
plt.show()
```

```python
Ns = [1, 2, 4, 10]

plt.figure(figsize=(12, 8))

for index, N in enumerate(Ns):
    x_means = []
    for _ in range(10000):
        xs = []
        for i in range(N):
            x = np.random.rand() # 一様乱数を生成
            xs.append(x)
        x_mean = np.mean(xs)
        x_means.append(x_mean)
    plt.subplot(2, 2, index+1)
    plt.hist(x_means, bins=100, range=(0, 1), density=True)
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.title(f"N = {N}")
    plt.xlim(-0.05, 1.05)
    plt.ylim(0, 5)

plt.tight_layout()
plt.show()
```

## 1.4 サンプル和の確率分布

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_sums = []
N = 5

for _ in range(10000):
    xs = []
    for i in range(N):
        x = np.random.rand() # 一様乱数を生成
        xs.append(x)
    x_sum = np.sum(xs)
    x_sums.append(x_sum)

def normal(x, mu=0, sigma=1):
    return 1/(np.sqrt(2*np.pi)*sigma) * np.exp(-0.5*((x-mu)**2) / sigma**2)

x_norm = np.linspace(-5, 5, 1000)
mu = N/2
sigma = np.sqrt(N/12)
y_norm = normal(x_norm, mu=mu, sigma=sigma)

plt.hist(x_sums, bins=100, density=True)
plt.plot(x_norm, y_norm, c='r')
plt.xlim(-1, 6)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title(f"N = {N}")
plt.show()
```
